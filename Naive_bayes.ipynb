```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mlxtend.plotting import plot_decision_regions
import itertools
import matplotlib.gridspec as gridspec
from numpy import asarray
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB 
import matplotlib.pyplot as plt
from sklearn import datasets, neighbors

```

```python
fig = plt.figure(figsize=(12,8))
gs = gridspec.GridSpec(2, 2)
```

```python
data = pd.read_csv('/Users/Tanvi/Desktop/Tanvi_GitHub/naive_bayes.csv')
```

```python
data.head()
```

```python
X= data.loc[:,'x':'y']
Y = data.loc[:,'label']
```

```python
clf1 = LogisticRegression(random_state=1,solver='newton-cg',multi_class='multinomial')
clf2 = RandomForestClassifier(random_state=1, n_estimators=100)
clf3 = GaussianNB()
clf4 = neighbors.KNeighborsClassifier(n_neighbors=5)

```

```python
x1 = (data.iloc[:,0:2]) 
x1 = asarray(X)
y1 = (data.loc[:,'label'])
y1 = asarray(Y)
```

```python
fig = plt.figure(figsize=(10,8))

labels = ['Multinomial Logistic Regression', 'Random Forest', 'Naive Bayes', 'KNN']
for clf, lab, grd in zip([clf1, clf2, clf3, clf4],
                         labels,
                         itertools.product([0, 1], repeat=2)):

    clf.fit(x1, y1)
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=x1, y=y1, clf=clf, legend=2)
    plt.title(lab)

plt.show()


```
